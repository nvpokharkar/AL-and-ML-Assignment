Q.4) Explore the distribution of sales transactions over time by plotting a 
time series of transaction counts. 

```python
# Convert 'created_at' column to datetime
df_transactions['created_at'] = pd.to_datetime(df_transactions['created_at'])


# Extract date from 'created_at' and count transactions per day
transaction_counts = df_transactions.groupby(df_transactions['created_at'].dt.date).size()
transaction_counts
```

```python
# Plot time series of transaction counts

plt.figure(figsize=(10, 6))
transaction_counts.plot(color='g')
plt.title('Transaction Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

The graph between Date and Transaction count shows a gradual increase in transactions over time.

Q.8 Analyze customer behavior by examining the distribution of transaction counts per customer and identifying the most frequent customers

```python
# Group transactions dataframe by customer_id and count the number of transactions
transaction_counts = df_transactions.groupby('customer_id').size().reset_index(name='transaction_count')

# Merge transaction counts with customer details
customer_behavior = pd.merge(transaction_counts, df_customer, on='customer_id', how='left')

# Sort customers by transaction count in descending order
customer_behavior_sorted = customer_behavior.sort_values(by='transaction_count', ascending=False)

# Print the top 10 most frequent customers along with their details
print("Top 10 most frequent customers:")
customer_behavior_sorted[["customer_id", "first_name", "last_name", "email", "gender", "transaction_count"]].reset_index().head(10)
```

In this table, we see the top ten customers and their details with transaction counts.

```python
# Plot the distribution of transaction counts
transaction_counts['transaction_count'].plot(kind='hist', bins=65, color='Brown', edgecolor='black', figsize=(10,6))
plt.title('Distribution of Transaction Counts per Customer')
plt.xlabel('Transaction Count')
plt.ylabel('Frequency')
plt.show()
```

The x-axis shows transaction count, while the y-axis indicates frequency. The red line shows the average transaction count.
The distribution tends to the right, indicating that most consumers have a low volume of transactions. There are a few consumers with a large number of transactions.

Q.9) Calculate summary statistics for relevant variables (e.g., total sales revenue, average transaction size) to understand the company's sales performance better.

```python
# Calculate total sales revenue.

total_revenue = df_transactions['total_amount'].sum()

# Calculate average transaction size

average_transaction_size = df_transactions['total_amount'].mean()

# Display summary statistics.

print("Summary Statistics:")
print(f"Total Sales Revenue: ${total_revenue:.2f}")
print(f"Average Transaction Size: ${average_transaction_size:.2f}")
print()
```

```python
# Calculate the other additional matrix like total sales revenue, average transaction size.

max_transaction_size = df_transactions['total_amount'].max()
min_transaction_size = df_transactions['total_amount'].min()

# Print additional summary statistics if needed
print("Additional summary statistics: ")
print(f"Maximum Transaction Size: ${max_transaction_size:.2f}")
print(f"Minimum Transaction Size: ${min_transaction_size:.2f}")
```

Q.7)Explore the distribution of sales transactions by product and identify 
the top-selling products. 

```python
df_merged = pd.merge(left=df_transactions, right=df_product, left_on=df_transactions.product_id, right_on=df_product.id, how="inner" )
product_sales = df_merged.groupby(['product_id', 'productDisplayName']).agg({'quantity': 'sum'}).reset_index()

# Sort products by quantity sold in descending order to identify top-selling products using product Display name.
top_selling_products = product_sales.sort_values(by='quantity', ascending=False).head(10)

# Fetch details of top-selling products from products dataframe
top_selling_products_details = pd.merge(top_selling_products, df_product, left_on='product_id', right_on='id', how='left')

# top-selling products using product Display name.


print("Top-selling products:")
top_selling_products_details
```

Above information, I obtained the top ten selling products and their information, such as product name, quantity, product id, and so on, so that we can determine which product is most in demand.
